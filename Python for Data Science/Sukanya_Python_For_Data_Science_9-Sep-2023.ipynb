## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
ECom_Data['revenue1'] = ECom_Data[(ECom_Data['discounted_price'] > 600)]['discounted_price']*0.25
ECom_Data['revenue2'] = ECom_Data[(ECom_Data['discounted_price']>350) & (ECom_Data['discounted_price']<=600)]['discounted_price']*0.15
ECom_Data['revenue3'] = ECom_Data[(ECom_Data['discounted_price']>100) & (ECom_Data['discounted_price']<=350)]['discounted_price']*0.10
ECom_Data['revenue4'] = ECom_Data[(ECom_Data['discounted_price']<=100)]['discounted_price']*0.05

ECom_Data['revenue'] = ECom_Data['revenue1'].fillna(ECom_Data['revenue2']).fillna(ECom_Data['revenue3']).fillna(ECom_Data['revenue4'])
ECom_Data['revenue'].sum()
```

Total Revenue generated by the E-Commerce company over all orders placed = 2217486.85

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'])
ECom_Data['Month_Year'] = ECom_Data['Order_Date'].apply(lambda x: x.strftime("%Y %m"))
ECom_Data['Month'] = ECom_Data['Order_Date'].dt.month
ECom_Data['Year'] = ECom_Data['Order_Date'].dt.year
ECom_Data

```

```python
df = ECom_Data.groupby(['Region','Month_Year']).revenue.sum().reset_index()
df.head()
```

```python
sns.lineplot(df, x='Month_Year', y = 'revenue', hue="Region")
plt.title("Monthly Revenue for Each Region")
plt.xticks(rotation=90)
plt.show()
```
