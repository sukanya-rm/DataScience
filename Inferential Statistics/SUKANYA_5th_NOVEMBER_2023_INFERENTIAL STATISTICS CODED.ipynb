# Problem 3
Zingaro stone printing is a company that specializes in printing images or patterns on polished or unpolished stones. However, for the optimum level of printing of the image, the stone surface has to have a Brinell's hardness index of at least 150. Recently, Zingaro has received a batch of polished and unpolished stones from its clients. Use the data provided to answer the following (assuming a 5% significance level);

# 3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

## 3.1.1 State the Null and Alternate Hypothesis.

### H0: The null hypothesis is that the unpolished stones are suitable for printing. 

### HA: The alternative hypothesis is that the unpolished stones may not be suitable for printing.


### The level of significance, alpha = 0.05

## 3.1.2 Conduct the hypothesis test and compute the p_value.     


### We will be performing the One-sample T-test.

```python
from scipy.stats import ttest_1samp
import statsmodels.stats.api as sm

df = pd.read_csv('Zingaro_company.csv')
df.head()
```

```python
print("Sample size of data is", len(df))
```

Here we select $\alpha$ = 0.05 and the population standard deviation is not known.

#Identify the test statistic
* We have one sample for unpolished stones and we do not know the population standard deviation.
* The sample size is n > 30. So we use 1-sample T-test.

```python
#Calculate the p_value and the test_statistic

t_statistic, p_value = stats.ttest_1samp(df['Unpolished '],150)
print("ONE-SAMPLED T-TEST")
print("The t_statistic is", t_statistic)
print("The p_value is ", p_value)
```

```python
alpha_level = 0.05

if p_value < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis.')
    print('We conclude that the unpolished stones may not be suitable for printing.')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis.')
    print('We conclude that the unpolished stones are suitable for printing.')
```

## 3.1.3 Write down conclusions from the test results.   


### As the p_value is less than the level of significance, we can reject the null hypothess that the unpolished stones are suitable for printing. 

### Zingaro is justified in thinking that unpolished stones may not be suitable for printing. As the p_value is less than the level of significance, we can reject the null hypothesis. 

# 3.2 Is the mean hardness of the polished and unpolished stones the same?

## 3.2.1 State the null and alternate hypotheses. 


### The null hypothesis is that the mean hardness of the polished and the unpolished stones is the same. 
### * $H_0$: $\mu{P}$ - $\mu{U}$ =      0    or    $\mu{P}$ = $\mu{U}$


### The alternative hypothesis is that the mean hardness of the polished and unpolished stones is not the same.
### * $H_A$: $\mu{P}$ - $\mu{U}$ $\neq$  0    or    $\mu{P}$ $\neq$ $\mu{U}$


## 3.2.2 Conduct the hypothesis test.

### Identify the test statistic

* We have two samples and we do not know the population standard deviation.
* Sample sizes for both samples are the same.
* The sample size is n > 30. So we perform the 2-Sample Independent T-test.

```python
#Calculate the p_value and the test_statistic

from scipy.stats import ttest_ind
import statsmodels.stats.api as sm

t_statistic, p_value = stats.ttest_ind(df['Unpolished '],df['Treated and Polished'])
print("TWO-SAMPLED INDEPENDENT T-TEST")
print("The t_statistic is", t_statistic)
print("The p_value is ", p_value)
```

```python
alpha_level = 0.05

if p_value < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis.')
    print('We conclude that the mean hardness of the polished stones and unpolished stones are not the same. ')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis.')
    print('We conclude that the mean hardness of the polished stones and unpolished stones are the same.')
```

## 3.2.3 Write down conclusions from the test results.   



### The null hypothesis has been rejected as p_value is less than alpha. We go with the alternate hypothesis. 

### Thus, the mean hardness of polished and unpolished stones is not the same.

In order to identify for which method mean Response is different from other groups,the hypotheses may be stated as:
0: All pairs of group means are equal against 
: At least one group mean is different from the rest.

0: 1 = 2 and 1 = 3 and 2 = 3 against 
: 1  2 or 1  3 or 2  3

where 1 represents mean response for method1, 2 represents mean response for method2 and 3 is the same for method3.

```python
# Posthoc test: Tukey test for alloy1

MultiComp=MultiComparison(df_dental_alloy1['Response'],df_dental_alloy1['Method'])
print(MultiComp.tukeyhsd().summary())
```

### Method1 and Method2: p_value > alpha, H0 is true. 1 = 2
### Method1 and Method3: p_value < alpha, H1 is true. 1  3
### Method2 and Method3: p_value < alpha, H1 is true. 2  3

### P-value is significant for comparing response for the pair Method1 and Method3 and for Method2 and Method3, but not for Method1 and Method2. The null hypothesis of equality of all population means is rejected. 

### It is now clear that mean response for Method1 and Method2 are similar but response for Method1 and Method3 as well as Method2 and Method3 are significantly different from Method1 and Method2.
